# [605. Can Place Flowers](https://leetcode.com/problems/can-place-flowers/description/)

You have a long flowerbed in which some of the plots are planted, and some are not. However, flowers cannot be planted in **adjacent**  plots.

Given an integer array `flowerbed` containing `0`'s and `1`'s, where `0` means empty and `1` means not empty, and an integer `n`, return `true`if `n` new flowers can be planted in the `flowerbed` without violating the no-adjacent-flowers rule and `false` otherwise.

**Example 1:** 

```
Input: flowerbed = [1,0,0,0,1], n = 1
Output: true
```

**Example 2:** 

```
Input: flowerbed = [1,0,0,0,1], n = 2
Output: false
```

**Constraints:** 

- `1 <= flowerbed.length <= 2 * 10^4`
- `flowerbed[i]` is `0` or `1`.
- There are no two adjacent flowers in `flowerbed`.
- `0 <= n <= flowerbed.length`

### Cases
- n = 0, return True
- Single element:
    - if 1, return false
    - if 0, return true
- n is more than possible plantation [eg: n=2 , arr[1,0,0,0,1]] or not enough non adjacent flowers possible [eg: n=2, arr[1,0,0,0,1,0]]
- adjacency:
    - middle element (checking left and right == 0 )
    - Start element (checking right == 0)
    - last element (checking left = 0)

### Pseudo Code
```
1.  Given n
2.      Traversing through the array
3.          If index = 0 and element = 0 and right element = 0
                initialize the element with '1'
4.              n -=1
5.          else if index = arr.length-1 and element = 0 and left element = 0
6.              Initialize the element to '1'
7.              n -=1
8.          else if element = 0 and its left and right elements are '0'
9.              Initialize the element to '1'
11.             n -=1
12.          If n==0 then return true
13.      if (n>0) then return false
```

### Approach:
- BruteForce
### Time Complexity:
- O(n)




### My Code

````python
class Solution:
    def canPlaceFlowers(self, flowerbed: List[int], n: int) -> bool:
        if n == 0: return True

        # For single element array
        if len(flowerbed) == 1 and flowerbed[0] == 1:
            return False
        elif len(flowerbed) == 1 and flowerbed[0] == 0:
            return True

        # For multiple element array
        for i in range(len(flowerbed)):
            if flowerbed[i] == 0:
                if i == 0 and flowerbed[i+1] == 0: #first element
                    flowerbed[i] = 1
                    n -= 1
                elif i == len(flowerbed)-1 and flowerbed[i-1] == 0: #last element
                    flowerbed[i] = 1
                    n -= 1
                elif flowerbed[i-1] == 0 and flowerbed[i+1] == 0: #middle element
                    flowerbed[i] = 1
                    n -= 1
                if n == 0: return True # if n flowers are planted already
        if n > 0: return False # insuffient space
````
